```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset into a DataFrame
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/HR_comma_sep.csv')  # Adjust the file path as needed

# Rename columns to match the provided headers if necessary
df.columns = ['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours',
              'time_spend_company', 'Work_accident', 'left', 'promotion_last_5years', 'sales', 'salary']

# 1. Number of employees in each department (sales)
num_employees_per_department = df.groupby('sales').size().reset_index(name='num_employees')
print("1. Number of employees in each department:")
print(num_employees_per_department)

# 2. Number of employees who left, department wise, and ranking
left_employees_per_department = df[df['left'] == 1].groupby('sales').size().reset_index(name='num_employees')
left_employees_per_department['rank'] = left_employees_per_department['num_employees'].rank(ascending=False).astype(int)
print("\n2. Number of employees who left, department wise, and ranking:")
print(left_employees_per_department)

# 3. Average time_spend_company and average average_montly_hours for employees left and not left under each department
avg_time_and_hours = df.groupby(['sales', 'left']).agg(
    avg_time_spend_company=('time_spend_company', 'mean'),
    avg_average_monthly_hours=('average_montly_hours', 'mean')
).reset_index().sort_values(by='sales')
print("\n3. Average time_spend_company and average average_montly_hours for left and not left under each department:")
print(avg_time_and_hours)

# 4. Minimum and maximum time_spend_company and average_montly_hours for left and not left employees under each department
min_max_time_and_hours = df.groupby(['sales', 'left']).agg(
    min_time_spend_company=('time_spend_company', 'min'),
    max_time_spend_company=('time_spend_company', 'max'),
    min_average_monthly_hours=('average_montly_hours', 'min'),
    max_average_monthly_hours=('average_montly_hours', 'max')
).reset_index().sort_values(by='sales')
print("\n4. Minimum and maximum time_spend_company and average_montly_hours for left and not left under each department:")
print(min_max_time_and_hours)

# 5. Minimum, maximum, and average last_evaluation and satisfaction_level for left and not left employees under each department
evaluation_satisfaction_stats = df.groupby(['sales', 'left']).agg(
    min_last_evaluation=('last_evaluation', 'min'),
    max_last_evaluation=('last_evaluation', 'max'),
    avg_last_evaluation=('last_evaluation', 'mean'),
    min_satisfaction_level=('satisfaction_level', 'min'),
    max_satisfaction_level=('satisfaction_level', 'max'),
    avg_satisfaction_level=('satisfaction_level', 'mean')
).reset_index().sort_values(by='sales')
print("\n5. Minimum, maximum, and average last_evaluation and satisfaction_level for left and not left under each department:")
print(evaluation_satisfaction_stats)

# 6. Total count of work_accidents and promotion_last_5years for left and not left employees under each department
work_accident_promotion_stats = df.groupby(['sales', 'left']).agg(
    count_work_accident=('Work_accident', 'sum'),
    count_promotion_last_5years=('promotion_last_5years', 'sum')
).reset_index().sort_values(by='sales')
print("\n6. Total count of work_accidents and promotion_last_5years for left and not left under each department:")
print(work_accident_promotion_stats)

# 7. Combine outputs from questions 3-6
combined_stats = pd.merge(avg_time_and_hours, min_max_time_and_hours, on=['sales', 'left'])
combined_stats = pd.merge(combined_stats, evaluation_satisfaction_stats, on=['sales', 'left'])
combined_stats = pd.merge(combined_stats, work_accident_promotion_stats, on=['sales', 'left'])
print("\n7. Combined statistics:")
print(combined_stats)

# 8. Plotting grouped bar chart for average time_spend_company
plt.figure(figsize=(12, 6))
sns.barplot(x='left', y='avg_time_spend_company', hue='sales', data=combined_stats, ci=None)
plt.title('Average Time Spent in Company by Employee Status and Department')
plt.xlabel('Employee Status (Left or Not Left)')
plt.ylabel('Average Time Spent in Company')
plt.legend(title='Department', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 9. Heatmap of the average monthly hours for left and not left employees
heatmap_data = df.pivot_table(values='average_montly_hours', index='sales', columns='left', aggfunc='mean')
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Average Monthly Hours for Left vs Not Left Employees by Department')
plt.tight_layout()
plt.show()
```
